```python
import networkx as nx    # network analysis, visualization library

gh = nx.Graph()
gh.add_node('n1')
gh.add_node('n2')
gh.add_node('n3')
gh.add_edge('n1','n2')
gh.add_edge('n1','n3')
gh.add_edge('n2','n3')

print(gh.nodes())
print(gh.edges())

nx.draw(gh)
```

```python
G=nx.Graph()
ls=[1,2,3]

G.add_nodes_from(ls)
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(2,3)

print(G.nodes())
print(G.edges())
nx.draw(G)
```
