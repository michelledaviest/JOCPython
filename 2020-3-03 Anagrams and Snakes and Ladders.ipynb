# Three ways to check two strings are annagrams or not

Given two strings find if it is an anagram
- tea and eat
- steam and teams
<br>

By sorting

```python
str1 = "steam"
str2 = "teams"
str1 = sorted(list(str1)) #you can also not convert to list sorted(str1)
str2 = sorted(list(str2))
print(str1 == str2)
```

By summing unicode(ord) values of the entire string

```python
sum1 = sum(map(lambda x: ord(x),list(str1)))
sum2 = sum(map(lambda x: ord(x),list(str2)))
print(sum1 == sum2)
```

By counting the number of characters in each word

```python
flag = True
if len(str1) != len(str2): flag = False
for ch in str1:
    if not flag: break
    if str1.count(ch) != str2.count(ch): flag = False
if flag: print("Anagrams!")
else: print("Not anagrams!")
```

# Snakes and Ladders using Pillow

```python
from PIL import Image
#path of image C:\\Users\\Michelle\\Documents\\Cummins\\Sem6\\JOCPython\\2020-3-03_docs

def show_br(): 
    board = Image.open("C:\\Users\\Michelle\\Documents\\Cummins\\Sem6\\JOCPython\\2020-3-03_docs\\SnakeLadder.jpg")
    board.show()

snakes = {
    44 : 22,
    46 : 5,
    48 : 9,
    52 : 11,
    55 : 7,
    59 : 17,
    64 : 36,
    69 : 33,
    73 : 1,
    83 : 19,
    92 : 51,
    95 : 24,
    98 : 28
}

ladders = {
    8 : 26,
    21 : 82, 
    43 : 77,
    50 : 91,
    54 : 93,
    62 : 96,
    66 : 87,
    80 : 100
}
```

```python
import random 

def roll_dice():
    return random.randint(1,6)

flag = True
score1, score2 = 0

show_br()

def check_play(score):
    if ladders.get(score) == None:
        if snakes.get(score) == None: 
            print("Score is",score)
        else:
            score = snakes.get(score)
            print("Snake! Your score is",score)
    else:
        score = ladders.get(score)
        print("Ladders! Your score is",score)
    return score
            
while score1<101 and score2<101:
    play = roll_dice()
    if flag:
        print("Now it is Player 1's turn")
        score1 += play
        if score1>100: break
        score1 = check_play(score1)
        flag = False
    else:
        print("Now it is Player 2's turn")
        score2 += play
        if score2>100: break
        score2 = check_play(score2)
        flag = True

if score1>score2: print("Player1 has won!")
else: print("Player2 has won!")
```
